Changes:
- set $\mu = 3\sigma_c$ in data
- renormalized averages. They were $\frac{1}{S}\sum N$, now are $\frac{1}{S^*} \sum N$, achieved by dividing by $\phi_N$

Allesina calculation:
$r_i = (K) - m + \mu K + z_i$
where the variance of $z_i$ is $\sigma_c^2 K^2$. We thus have 
$\phi_N = {\rm sf}[(K-m + \mu K)/(\sigma_c K)]$
where sf is the surival fraction in a normal distribution (1 - cumulative distribution)

```python
#Identity = pd.read_csv('Data/Community_gaussian_identity_quadratic_CVXOPT_log_v1.csv')
#Null = pd.read_csv('Data/Community_gaussian_null_quadratic_CVXOPT_log_v1.csv')
Identity = pd.read_csv('Data/Community_gaussian_identity_quadratic_CVXOPT_log_RAM.csv')
Null = pd.read_csv('Data/Community_gaussian_null_quadratic_CVXOPT_log_RAM.csv')
Data = [Identity,Null]
Legend = ["Identity","Random"]
fig, axs = plt.subplots(1, 3)
axs = axs.ravel()
K = 10
m = 1
for i in range(3):
    ax =axs[i]
    ax.set_xlim([10**(-6), 10**(3)])
    ax.set_xlabel(r'$\sigma_c$',fontsize=20)
    ax.set_xscale('log')
    ax.xaxis.set_tick_params(labelsize=20)
    ax.yaxis.set_tick_params(labelsize=20)
    for j in range(2):
        data=Data[j]
        #data =data[data['mu'] == 0.0] 
        if i==0:
            ax.plot(data['epsilon'],data['phi_N'],linestyle='dashed', alpha=0.7, label=Legend[j])
            ax.scatter(data['epsilon'], data['phi_N'], alpha=0.7, label='_nolegend_', s=15)
            ax.set_ylim([0, 1.1])
            ax.set_ylabel(r'$S^*/M$',fontsize=20)
            #ax.set_title(r'Self-renewing dynamics',fontsize=15)
        if i==1:
            ax.plot(data['epsilon'],data['mean_N']/data['phi_N'],linestyle='dashed', alpha=0.7, label=Legend[j])
            ax.scatter(data['epsilon'], data['mean_N']/data['phi_N'], alpha=0.7, label='_nolegend_', s=15)
            ax.set_ylim([0, 60.1])
            ax.set_ylabel(r'$\left<N\right>$',fontsize=20)
            #ax.set_title(r'Self-renewing dynamics',fontsize=15)
            #ax.legend(loc='lower left',bbox_to_anchor=(0.01, 0.17, 0.6, 0.6))
        if i==2:
            ax.plot(data['epsilon'],data['q_N']/data['phi_N'],linestyle='dashed', alpha=0.7, label=Legend[j])
            ax.scatter(data['epsilon'], data['q_N']/data['phi_N'], alpha=0.7, label='_nolegend_', s=15)
            ax.set_ylim([1,400.1])
            ax.set_ylabel(r'$\left<N^2\right>$',fontsize=20)
            #ax.set_title(r'Self-renewing dynamics',fontsize=15)
            #ax.legend(loc='lower left',bbox_to_anchor=(0.01, 0.17, 0.6, 0.6))

axs[0].plot(Data[0]['epsilon'],norm.sf((m-K*(1+Data[0]['mu']))/(Data[0]['epsilon']*K)),'k',label='No interactions')
axs[0].plot(Data[1]['epsilon'],norm.sf((m-K*(Data[1]['mu']))/(Data[1]['epsilon']*K)),'k',label='No interactions')
#axs[1].plot(Data[0]['epsilon'],[K*Data[0]['epsilon'].loc[i]*norm.expect(lambda x: x/((1+Data[0]['epsilon'].loc[i]**2)*K*Data[0]['epsilon'].loc[i]),loc=(K*(1+Data[0]['mu'].loc[i])-m)/(K*Data[0]['epsilon'].loc[i]),lb=0)/(norm.expect(lambda x: 1, loc=(K*(1+Data[0]['mu'].loc[i])-m)/(K*Data[0]['epsilon'].loc[i]),lb=0)+1e-10) for i in range(60)],'k',label='No interactions')
#axs[1].plot(Data[0]['epsilon'],norm.sf((m-K*(Data[0]['mu']))/(Data[0]['epsilon']*K)),'k',label='No interactions')
axs[1].legend(loc='lower left',bbox_to_anchor=(0.01, 0.13, 0.6, 0.6),prop={'size':20})
fig.set_size_inches(16, 5)
fig.tight_layout()
fig_name='community_properties_quadratic_mu3sig.pdf'

fig.savefig(fig_name, dpi=300, transparent=True)
```
