## Spectrums

### Pollination

```python
mypath=os.path.join(os.getcwd(), "Pollination")
#mypath=os.path.join(os.getcwd(), "Plant-Herbivore")
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for item in ["references.csv", "README"]:
    onlyfiles.remove(item)
for f in onlyfiles:
    if f.endswith(".csv"):
        fig, ax = plt.subplots(1,1,figsize=(6,5))
        df=pd.read_csv(os.path.join(mypath, f), index_col=0)
        df=df.drop(['Numbers of flowers'], axis=1, errors='ignore')
        C=np.asarray(df).transpose()
        print(f,C.shape)
        ls=np.real(LA.eigvals(np.dot(C,C.transpose())))
        ax.xaxis.set_tick_params(labelsize=20)
        ax.yaxis.set_tick_params(labelsize=20)
        ax.set_title(f,fontsize=20)
        n_bins=130
        ax.hist(ls,bins =np.linspace(0,20.1, n_bins),density=1, color="salmon")
        fig.tight_layout()
        fig_name=mypath+'/'+f+'.pdf'
        fig.savefig(fig_name, dpi=300, transparent=True)
```
